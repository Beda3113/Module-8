```python
import sympy as sp

x = sp.symbols('x')

functions = {
    'a': x**2 - 30*x + 1,
    'b': -x**3 + 5*x - 7*x**2 + 17,
    'c': x**2 - 1/x,
    'd': 1/x**5 + x**17 + 8
}

def find_extrema(func):
    first_derivative = sp.diff(func, x)
    critical_points = sp.solve(first_derivative, x)
    extrema_info = {}

    for point in critical_points:
        if point.is_real:
            second_derivative = sp.diff(first_derivative, x)
            concavity = second_derivative.subs(x, point)
            value_at_point = func.subs(x, point)

            extrema_info[point] = {
                'value': value_at_point,
                'type': 'minimum' if concavity > 0 else 'maximum' if concavity < 0 else 'inflexion'
            }

    return extrema_info

for key, func in functions.items():
    extrema = find_extrema(func)
    print(f"Function {key}:")
    print(f"Extrema: {extrema}\n")
```

```python
import sympy as sp

# Определение переменной
x = sp.symbols('x')

# Определение функций
functions = {
    'a': x**2 - 30*x + 1,
    'b': -x**3 + 5*x - 7*x**2 + 17,
    'c': x**2 - 1/x,
    'd': 1/x**5 + x**17 + 8
}

def newton_method(f, x_start, epsilon, num_iterations):
    x_curr = x_start
    df_x = sp.diff(f, x)  # первая производная
    d2f_x = sp.diff(df_x, x)  # вторая производная

    for i in range(num_iterations):
        dfx_val = df_x.subs(x, x_curr)
        d2fx_val = d2f_x.subs(x, x_curr)

        if dfx_val == 0:
            print("Первая производная равна нулю. Метод Ньютона не может быть применен.")
            return None, i

        if d2fx_val == 0:
            print("Вторая производная равна нулю. Метод Ньютона не может быть применен.")
            return None, i

        x_new = x_curr - dfx_val / d2fx_val

        if abs(x_new - x_curr) < epsilon:
            return float(x_new.evalf()), i + 1  # Возвращаем количество итераций

        x_curr = x_new

    return float(x_curr.evalf()), num_iterations  # Возвращаем количество итераций

def gradient_descent(f, x_start, learning_rate, epsilon, max_iterations):
    x_curr = x_start
    df_x = sp.diff(f, x)  # первая производная

    for i in range(max_iterations):
        dfx_val = df_x.subs(x, x_curr)

        if abs(dfx_val) < epsilon:
            return float(x_curr), i + 1  # Возвращаем количество итераций

        x_new = x_curr - learning_rate * dfx_val

        if abs(x_new - x_curr) < epsilon:
            return float(x_new), i + 1  # Возвращаем количество итераций

        x_curr = x_new

    return float(x_curr), max_iterations  # Возвращаем количество итераций

def compare_methods(func, x0_newton, x0_grad, learning_rate, epsilon, max_iterations):
    true_minimum = sp.solve(sp.diff(func, x), x)  # Находим истинный минимум
    true_minimum_value = func.subs(x, true_minimum[0]).evalf() if true_minimum else None

    # Метод Ньютона
    min_newton, iterations_newton = newton_method(func, x0_newton, epsilon, max_iterations)

    # Градиентный спуск
    min_gradient, iterations_gradient = gradient_descent(func, x0_grad, learning_rate, epsilon, max_iterations)

    print(f"Функция: {func}")

    if min_newton is not None:
        print(f"Метод Ньютона: {min_newton}, Итерации: {iterations_newton}, Ошибка: {abs(min_newton - true_minimum_value)}")
    else:
        print(f"Метод Ньютона: Не удалось найти минимум за {iterations_newton} итераций.")

    print(f"Градиентный спуск: {min_gradient}, Итерации: {iterations_gradient}, Ошибка: {abs(min_gradient - true_minimum_value) if true_minimum_value is not None else 'Неизвестно'}")
    print("-" * 50)

# Параметры
epsilon = 1e-5
max_iterations = 100
learning_rate = 0.01

# Начальные точки для каждого метода
initial_points = {
    'a': (15, 15),
    'b': (1, 1),
    'c': (2, 2),  # Начальная точка для метода Ньютона и градиентного спуска
    'd': (1, 1)
}

for func_name, func in functions.items():
    x0_newton, x0_grad = initial_points[func_name]
    compare_methods(func, x0_newton, x0_grad, learning_rate, epsilon, max_iterations)


```
